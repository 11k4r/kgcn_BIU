```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('final_results.csv', names=['dataset', 'aggregator', 'parameter', 'parameter_value', 'auc', 'f1'])
```

```python
results.groupby('aggregator')[['auc', 'f1']].agg(np.max)
```

```python
x = results[results['parameter'] == 'k']['parameter_value'].unique()
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'k')]['auc'])
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'k')]['f1'])
plt.legend(['auc', 'f1'])
plt.title("K")
```

```python
x = results[results['parameter'] == 'dim']['parameter_value'].unique()
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'dim')]['auc'])
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'dim')]['f1'])
plt.legend(['auc', 'f1'])
plt.title("dim")
```

```python
x = results[results['parameter'] == 'h']['parameter_value'].unique()
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'h')]['auc'])
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'h')]['f1'])
plt.legend(['auc', 'f1'])
plt.title("H")
```

```python
x = results[results['parameter'] == 'h']['parameter_value'].unique()
plt.plot(x, results[(results['aggregator'] == 'neighbor') & (results['parameter'] == 'h')]['auc'])
plt.plot(x, results[(results['aggregator'] == 'concat') & (results['parameter'] == 'h')]['auc'])
plt.plot(x, results[(results['aggregator'] == 'sum') & (results['parameter'] == 'h')]['auc'])
plt.legend(['neighbor', 'concat', 'sum'])
plt.title("H")
```
